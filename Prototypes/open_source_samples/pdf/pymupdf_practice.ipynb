```python
for i in range(doc.page_count):
    image_list = doc.load_page(i).get_images()
    for image in image_list:
        xref = image[0]
        pix = pymupdf.Pixmap(doc, xref)
        if pix.n < 5:
            pix._writeIMG(f'images/{xref}.png', format_='png', jpg_quality=None)
        else:
            pix1 = pymupdf.open(pymupdf.csRGB, pix)
            pix._writeIMG(f'images/{xref}.png', format_='png', jpg_quality=None)
        pix1 = None
        
```

### Trying Tabula for pdf extraction

```python
import pandas as pd
import tabula

# Try these encoding options
encodings = ['latin-1', 'iso-8859-1', 'cp1252']

for encoding in encodings:
    try:
        dfs = tabula.read_pdf(pdf_path, 
                              pages='all', 
                              multiple_tables=True, 
                              encoding=encoding)
        break
    except UnicodeDecodeError:
        continue

for i, df in enumerate(dfs):
    df.to_csv(f"Tables/table_{i}.csv")
#     print(df)
```
